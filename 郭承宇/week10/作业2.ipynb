```python
import base64
from openai import OpenAI
import fitz

client = OpenAI(
    api_key='sk-e0e141c7108a4cabaf4cced1f749ae89',
    base_url="https://dashscope.aliyuncs.com/compatible-mode/v1"
)

pdf_path = 'D:\\笔记软件\\Obsidian仓库\\计算机基础\\MySQL\\SQLCheatSheet-ByGeekHour.pdf'

question = "请将PDF中的内容进行总结"

# 由于阿里云的服务器无法访问本地电脑，它只能访问公网链接或者base64编码的图片数据，故需要下面的函数
# PDF 转图片 再转Base64
def pdf_to_base64(pdf_file_path,page_num=0):
    doc = fitz.open(pdf_file_path) # 打开PDF文件
    page = doc[page_num] # 获取指定页码的页面对象

    zoom = 2 # 将 PDF 放大2倍 PDF是矢量文件，无限放大都是清晰的
    mat = fitz.Matrix(zoom, zoom)

    pix = page.get_pixmap(matrix=mat) # 渲染页面为图片

    image_data = pix.tobytes("png")

    base64_data = base64.b64encode(image_data).decode('utf-8')
    return base64_data

img_base64 = pdf_to_base64(pdf_path)

completion = client.chat.completions.create(
    model='qwen3-vl-plus',
    messages=[
        {
            'role': 'system',
            'content': '你是一个PDF解析专家，请仔细阅读图片内容'
        },
        {
            'role': 'user',
            'content': [
                {
                    'type': 'image_url',
                    'image_url': {
                        'url': f'data:image/png;base64,{img_base64}'
                    }
                },
                {
                    'type': 'text',
                    'text': question
                }
            ]
        }
    ],
    stream=True
)


for chunk in completion:
    if chunk.choices:
        content = chunk.choices[0].delta.content
        if content:
            print(content, end='')
```
